```python
import pandas as pd
dfl= pd.DataFrame({'X': [10, 0], 'Y': [5, 5]})
df2= pd.DataFrame({'X': [2, 2], 'Y': [4, 4]})
take_smaller = lambda s1, s2: s1 if s1.sum() < s2.sum() else s2
dfl.combine(df2, take_smaller)
print(dfl)
```
